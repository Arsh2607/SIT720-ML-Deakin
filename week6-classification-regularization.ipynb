```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/house-prices/train.csv") # from https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data
print(df.columns)
```

```python

x = df['GrLivArea']
y = df['SalePrice']
# Calculate the slope and intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

# Print the weights
print("Slope: ", slope)
print("Intercept: ", intercept)

# Create a scatter plot of the data points
plt.scatter(x, y)

# Add the regression line to the plot
plt.plot(x, slope * x + intercept, color='red')

# Set the title and labels for the plot
plt.title('Regression Line Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.show()
```

```python
# Calculate the mean of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate the variance of x and the covariance of x and y
variance_x = np.sum((x - x_mean)**2)
covariance_xy = np.sum((x - x_mean) * (y - y_mean))

# Calculate the slope and intercept of the regression line
slope = covariance_xy / variance_x
intercept = y_mean - slope * x_mean

# Print the weights
print("Slope: ", slope)
print("Intercept: ", intercept)
```
